```python
import pandas as pd
import numpy as np
import re
```

```python
# load progress.csv
df = pd.read_csv('raw.csv', header=None, names=['raw'])
df['gpu'] = df.iloc[:,0].apply(lambda x: re.findall("\d+", str(x))[0])
df['pid'] = df.iloc[:,0].apply(lambda x: re.findall("\d+", str(x))[1])
df['memory'] = df.iloc[:,0].apply(lambda x: re.findall("\d+", str(x))[2])
df
```

```python
# Save as pid.csv
df1 = pd.DataFrame(df, columns=['pid', 'gpu', 'memory'])
df1.to_csv('pid.csv', index=False)
df1
```
